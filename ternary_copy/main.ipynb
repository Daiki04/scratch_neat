```python
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots

df = pd.read_csv('result.csv')

global_fitness = df['best_fitness']
global_acc = df['best_acc']
global_test_fitness = df['best_test_fitness']
global_test_acc = df['best_test_acc']
fitness = df['fitness']
acc = df['acc']

plt.style.use(['science', 'grid', 'no-latex'])

fig, ax = plt.subplots(2, 1, figsize=(6, 8))

ax[0].plot(global_fitness, label='Global Best Fitness')
ax[0].plot(fitness, label='Local Best Fitness', linestyle='--')
ax[0].set_xlabel('Iteration')
ax[0].set_ylabel('Fitness')
ax[0].legend()

ax[1].plot(global_acc, label='Global Best Accuracy')
ax[1].plot(acc, label='Local Best Accuracy', linestyle='--')
ax[1].set_xlabel('Iteration')
ax[1].set_ylabel('Accuracy')
ax[1].legend()

plt.tight_layout()
plt.show()
```
