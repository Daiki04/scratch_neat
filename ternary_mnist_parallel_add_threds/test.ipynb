```python
import tensorflow as tf
import numpy as np
```

```python
def _create_dataset():
    mnist = tf.keras.datasets.mnist
    (x_train, y_train), (x_test, y_test) = mnist.load_data()
    x_train, x_test = x_train.reshape([-1, 784]).astype('float32') / 255, x_test.reshape([-1, 784]).astype('float32') / 255
    train_indices = np.where((y_train == 0) | (y_train == 1))[0]
    test_indices = np.where((y_test == 0) | (y_test == 1))[0]
    x_train = x_train[train_indices]
    y_train = y_train[train_indices]
    x_test = x_test[test_indices]
    y_test = y_test[test_indices]
    # データ数が多いので削減
    x_train = x_train[:1000]
    y_train = y_train[:1000]
    return (x_train, y_train), (x_test, y_test)
```

```python
(x_train, y_train), (x_test, y_test) = _create_dataset()
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
from concurrent.futures import ThreadPoolExecutor

class Test1:
    def __init__(self):
        self.pop = {i: Test2() for i in range(10)}

class Test2:
    def __init__(self):
        self.fit = None

def func1(a):
    with ThreadPoolExecutor(max_workers=4) as executor:
        for p in a.pop.values():
            executor.submit(func2, p)

def func2(a):
    a.fit = 1

b = Test1()
func1(b)
```
